```python
from googleapiclient.discovery import build
import pandas as pd
import http.client 
import os
import numpy as np
```

```python
api_key = "***"
youtube = build('youtube', 'v3', developerKey=api_key)
```

```python
def get_channel_videos(channel_id):
    res = youtube.channels().list(id=channel_id,part='contentDetails').execute()
    playlist_id = res['items'][0]['contentDetails']['relatedPlaylists']['uploads']
    
    videos = []
    next_page_token = None
    # loop through all pages to get the video list
    while 1:
        res = youtube.playlistItems().list(
            playlistId=playlist_id,
            part='snippet',
            maxResults=50,
            pageToken=next_page_token).execute()
        videos += res['items']
        next_page_token = res.get('nextPageToken')
        
        if next_page_token is None:
            break
    
    return videos
```

```python
#for i in range(0, len(video_ids), 50):
def get_videos_stats(video_ids):
    # you can get up to 50 per time, save up the API requests
    stats = []
    for i in range(0, 1000, 50):
        res = youtube.videos().list(id=','.join(video_ids[i:i+50]),part='statistics').execute()
        stats += res['items']
    return stats
```

```python
video_ids = list(map(lambda x:x['snippet']['resourceId']['videoId'], videos))
```

```python
stats = get_videos_stats(video_ids)
```

```python
#stats[0]
```

```python
# video id, title, likeCount, dislikeCount, viewCount, publishedAt
result = []
for video, stat in zip(videos, stats):
    temp = {
        "videoId": stat["id"],
        "title": video["snippet"]["title"],
        "channel Id": video["snippet"]["channelId"],
        "channel title": video["snippet"]["channelTitle"],
        "likeCount": stat["statistics"]["likeCount"],
        "dislikeCount": stat["statistics"]["dislikeCount"],
        "viewCount": stat["statistics"]["viewCount"],
        "publishedAt": video["snippet"]["publishedAt"]
    }
    result.append(temp)
```

```python
veritasium = pd.DataFrame(result)
#print (veritasium)
#print (veritasium['publishedAt'])
```

```python
veritasium['publishedAt'] = pd.to_datetime(veritasium['publishedAt'])  
```

```python
start_date = '2020-10-5'
end_date = '2020-10-11'

after_start_date = veritasium['publishedAt'] >= start_date
before_end_date = veritasium['publishedAt'] <= end_date
between_two_dates = after_start_date & before_end_date
filtered_dates = veritasium.loc[between_two_dates]
#print (filtered_dates)
```
